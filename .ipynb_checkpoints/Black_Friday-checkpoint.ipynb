# Hypothesis Test, Feature Engineering, & Model Selection

### Hypothesis Test

So let's test the hypothesis that women and men spend differently on Black Friday.  First let's do some permutation sampling.

```python
def permutation_sample(data1, data2):
    
    # Concatenate the data sets: data
    data = np.concatenate((data1,data2))

    # Permute the concatenated array: permuted_data
    permuted_data = np.random.permutation(data)

    # Split the permuted array into two: perm_sample_1, perm_sample_2
    perm_sample_1 = permuted_data[:len(data1)]
    perm_sample_2 = permuted_data[len(data1):]

    return perm_sample_1, perm_sample_2
```

```python
plt.figure(figsize=(20,10))
for _ in range(50):
    perm_sample_1, perm_sample_2 = permutation_sample(x, z)
    ecdf3 = ECDF(perm_sample_1)
    ecdf4 = ECDF(perm_sample_2)
    plt.plot(ecdf3.x,ecdf3.y, marker = '.', linestyle='none', color='hotpink', alpha=.005, markersize=2)
    plt.plot(ecdf4.x,ecdf4.y, marker = '.', linestyle='none', color='maroon', alpha=.005, markersize=2)
    
plt.plot(ecdf1.x,ecdf1.y, color='green', label = "Males")
plt.plot(ecdf2.x,ecdf2.y, color='blue', label = "Females")
plt.plot(ecdf.x,ecdf.y, color='red', label = "Combined")
plt.legend(loc='lower right', prop={'size':20})
plt.show()
```

And here we see that males are clearly outside of the 50 permutation samples, and the females might be as well.  Let's run a simple t-test to see if the two groups differ.

```python
x1, z1 = stats.ttest_ind(x,z)
print("t = " + str(x1))
print("p = " + str(2*z1))
```

So yeah, the two groups are very different in their Black Friday purchasing.

### Feature Engineering

But as is, I'm already running long probably so let's get to it. First, I need to dummy-ize my categorical variables and organize my dataframe into X and y dataframes. 

```python
df2 = pd.get_dummies(df1)
df2.head()
cols = list(df2.columns.values) #Make a list of all of the columns in the df
cols.pop(cols.index('Purchase')) #Remove Purchase from list
cols.pop(cols.index('PurchaseTx')) #Remove PurchaseTx from list
df3 = df2[cols+['PurchaseTx']] #Put PurchaseTx back in at the end
X = df3.iloc[:,:-1]
y = df3.iloc[:,-1]
```

So here I performed dummy encoding on my categorical variables and set up my X and y variables for the model. 

# Machine Learning Model Implementation

```python
# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)

# Create the regressor: reg_all
reg_all = LinearRegression()

# Fit the regressor to the training data
reg_all.fit(X_train, y_train)

# Predict on the test data: y_pred
y_pred = reg_all.predict(X_test)
cv_scores = cross_val_score(reg_all, X, y, cv=5)

# Print the 5-fold cross-validation scores
print(cv_scores)

print("Average 5-Fold CV Score: {}".format(np.mean(cv_scores)))

# Compute and print R^2 and RMSE
print("R^2: {}".format(reg_all.score(X_test, y_test)))
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error: {}".format(rmse))
```

So my model is... not good.  I'm explaining around 15% of the variance depending on what seed you get, and my cross validation score appears to be below even that. 

```python
steps = [('scaler', StandardScaler()),
        ('LR', LinearRegression())]
        
# Create the pipeline: pipeline
pipeline = Pipeline(steps)

# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y ,test_size=.3, random_state=42)
 
# Fit the pipeline to the training set: knn_scaled
LR_scaled = pipeline.fit(X_train,y_train)

# Instantiate and fit a k-NN classifier to the unscaled data
LR_unscaled = LinearRegression().fit(X_train, y_train)

# Compute and print metrics
print('Accuracy with Scaling: {}'.format(LR_scaled.score(X_test,y_test)))
print('Accuracy without Scaling: {}'.format(LR_unscaled.score(X_test,y_test)))
```

### Hyperparameter Adjustment

So let's look and see if we can tune this a little better.

```python
param_grid = {'n_neighbors': np.arange(1, 50)}
knn = KNeighborsClassifier()
knn_cv = GridSearchCV(knn, param_grid, cv=5)
knn_cv.fit(X, y)
print(knn_cv.best_params_)

print(knn_cv.best_score_)
```

The answer is, no, we can't.  Our best score (with a knn of 43) is barely better than flipping a coin. Let's look at the ROC curve and do a 5-fold cross-validation so we can see a better metric of how good our model is. 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

knn = KNeighborsClassifier(n_neighbors = 6)

# Fit the data
knn = knn.fit(X_train, y_train)

# Predict on the data
y_prediction = knn.predict(X_test)

y_pred_prob = knn.predict_proba(X_test)[:,1]

# Generate ROC curve values: fpr, tpr, thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

# Plot ROC curve
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label = 'KNN')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
print("AUC: {}".format(roc_auc_score(y_test, y_pred_prob)))

# Compute cross-validated AUC scores: cv_auc
cv_auc = cross_val_score(knn, X, y, cv=5, scoring='roc_auc')

# Print list of AUC scores
print("AUC scores computed using 5-fold cross-validation: {}".format(cv_auc))
```

So there you have it.  My AUC is around a .54 depending on the seed.  Overall, not great.
